## Show Distinct diamonds color

## What is the most common color grade?

```python
%%sql
SELECT color, COUNT(*) AS count
FROM diamonds
GROUP BY color
ORDER BY count DESC
LIMIT 1;

```

## How does the cut of a diamond affect its average price per carat?

```python
%%sql
SELECT cut, CAST(AVG(price/carat) AS decimal(10, 2)) AS avg_price_per_carat
FROM diamonds
GROUP BY cut
ORDER BY avg_price_per_carat DESC;
```

## Are there any specific cuts that consistently result in higher prices across different carat weights?

```python
%%sql
SELECT cut, sum(price) AS average_price
FROM diamonds
GROUP BY cut
HAVING COUNT(DISTINCT carat) > 1
ORDER BY average_price DESC;
```

## How does the depth percentage vary among different clarity grades

```python
%%sql
SELECT clarity, MIN(depth) AS min_depth, MAX(depth) AS max_depth, CAST(AVG(depth)  AS decimal(10, 2)) AS average_depth
FROM diamonds
GROUP BY clarity;
```

## How does the depth percentage affect diamonds price

```python
%%sql
SELECT 
  CASE 
    WHEN depth  <=43 THEN '43'
    WHEN depth >= 50.0 AND depth < 59.9 THEN '50-59%'
    WHEN depth >= 60.0 AND depth < 69.9 THEN '60-69'
    WHEN depth >= 70.0 AND depth < 79.9 THEN '70-79'
  END AS depth_range,
  CAST(AVG(price) AS decimal(10, 2)) AS average_price
FROM diamonds
GROUP BY depth_range
ORDER BY depth_range;
```

## How does the depth percentage vary across different diamond cuts?

```python
%%sql
SELECT cut, MIN(depth) AS min_depth, MAX(depth) AS max_depth, CAST(AVG(depth)AS decimal(10, 2))AS average_depth
FROM diamonds
GROUP BY cut;
```

## How does the depth percentage impact the average price per carat for diamonds of different color grades?


```python
%%sql
SELECT color, AVG(price / carat) AS average_price_per_carat, AVG(depth) AS average_depth_percentage
FROM diamonds
GROUP BY color
ORDER BY color;
```


## How does the table percentage affect the diamond price


```python
%%sql
SELECT
  CASE
    WHEN "table" >= 40 AND "table" <= 49.9 THEN '40-49.9'
    WHEN "table" >= 50 AND "table" <= 59.9 THEN '50-59.9'
    WHEN "table" >= 60 AND "table" < 69.9 THEN '60-69.9'
    WHEN "table" >= 70 AND "table" < 79.9 THEN '70-79.9'
  END AS table_range,
  COUNT(*) AS table_count,
  AVG(price) AS average_price
FROM diamonds
GROUP BY table_range
ORDER BY table_range;
```

## How does the table percentage impact the average price per carat for diamonds of different color grades?

```python
%%sql
SELECT
  CASE
    WHEN "table" >= 10 AND "table" <= 19.9 THEN '40-49.9'
    WHEN "table" >= 20 AND "table" <= 29.9 THEN '30-39.9'
    WHEN "table" >= 30 AND "table" <= 39.9 THEN '40-49.9'
    WHEN "table" >= 40 AND "table" <= 49.9 THEN '40-49.9'
    WHEN "table" >= 50 AND "table" <= 59.9 THEN '50-59.9'
    WHEN "table" >= 60 AND "table" < 69.9 THEN '60-69.9'
    WHEN "table" >= 70 AND "table" < 79.9 THEN '70-79.9'
  END AS table_range,
  color,
  AVG(price / carat) AS avg_price_per_carat
FROM diamonds
GROUP BY table_range, color
ORDER BY table_range, color;


```

## Do diamonds with a specific table percentage range tend to have more or fewer inclusions or flaws?

```python
%%sql
SELECT
  CASE
    WHEN "table" >= 10 AND "table" <= 19.9 THEN '40-49.9'
    WHEN "table" >= 20 AND "table" <= 29.9 THEN '30-39.9'
    WHEN "table" >= 30 AND "table" <= 39.9 THEN '40-49.9'
    WHEN "table" >= 40 AND "table" <= 49.9 THEN '40-49.9'
    WHEN "table" >= 50 AND "table" <= 59.9 THEN '50-59.9'
    WHEN "table" >= 60 AND "table" < 69.9 THEN '60-69.9'
    WHEN "table" >= 70 AND "table" < 79.9 THEN '70-79.9'
  END AS table_range,
  clarity,
  COUNT(*) AS inclusion_count
FROM diamonds
GROUP BY table_range, clarity
ORDER BY table_range, clarity;

```
